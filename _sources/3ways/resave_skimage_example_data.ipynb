```python
from skimage.data import cells3d
from skimage.io import imsave, imshow
```

```python
cells = cells3d()
cells.shape
```

```python
imshow(cells[30,0])
```

```python
membranes = cells[:,0]
membranes.shape
```

```python
imsave("../../data/cells3d_membranes.tif", membranes)
```

```python
nuclei = cells[:,1]
imshow(nuclei[30])
```

```python
imsave("../../data/cells3d_nuclei.tif", nuclei)
```
