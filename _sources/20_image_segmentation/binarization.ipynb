# Image Binarization

When binarizing an image, we produce an image that has only two values: True and False. They may also contain the value `0`, e.g. for background, and any other value for foreground.

```python
import numpy as np
from skimage.io import imread, imshow
from skimage.filters import gaussian
```

We use this example image of nuclei.

```python
image_nuclei = imread('../../data/mitosis_mod.tif')

imshow(image_nuclei)
```

## Image Thesholding
The most common binarization technique is thresholding. We _apply a threshold_ to determine which pixel lie above a certain pixel intensity and which are below.

```python
image_binary = image_nuclei > 60
```

```python
imshow(image_binary)
```

## Improving binarization results
Sometimes, binarization results appear pixelated. This can be improved by applying a filter before thresholding the image.

```python
image_denoised = gaussian(image_nuclei, sigma=1, preserve_range=True)
imshow(image_denoised, cmap='Greys_r')
```

```python
image_binary2 = image_denoised > 60

imshow(image_binary2)
```
