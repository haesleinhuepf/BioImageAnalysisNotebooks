# Equivalence testing
null-hypothesis: differences are out of a given range
alternative hypothesis: differences are small; and thus, bot samples are "equal" up to a given tolerance

```python
from numpy import random

x1 = random.normal(loc=5, scale=1, size=1000)
x2 = random.normal(loc=5, scale=1, size=1000)
```

```python
import matplotlib.pyplot as plt
plt.plot(x1, x2, ".")
plt.show()
```

```python
from statsmodels.stats.weightstats import ttost_ind

pval = ttost_ind(x1, x2, low=-0.2, upp=0.2)

pval
```

## Opposite situation

```python
from numpy import random

x1 = random.normal(loc=6, scale=1, size=1000)
x2 = random.normal(loc=5, scale=1, size=1000)
```

```python
import matplotlib.pyplot as plt
plt.plot(x1, x2, ".")
plt.show()
```

```python
from statsmodels.stats.weightstats import ttost_ind

pval = ttost_ind(x1, x2, low=-0.2, upp=0.2)

pval
```
